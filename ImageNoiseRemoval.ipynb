```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plots
%matplotlib inline
```

```python
# Function to display images side by side
def show_images(images, titles, cmap='gray', figsize=(15, 5)):
    plt.figure(figsize=figsize)
    for i in range(len(images)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(images[i], cmap=cmap)
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Read the noisy image
# You can change this to your image path
image_path = "E:\\FLIR_ADAS_1_3_REDUCED\\train\\thermal_8_bit\\FLIR_00016.jpeg" # Replace with your image file name
noisy_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
if noisy_img is None:
    raise FileNotFoundError("Image not found. Please check the path or name.")

#  Display the original noisy image
show_images([noisy_img], ["Original Noisy Image"])

# -------------------------------------------------------------
# 1. Averaging Filter (Mean Filter)
avg_filtered = cv2.blur(noisy_img, (5, 5))

# 2. Gaussian Blur
gaussian_filtered = cv2.GaussianBlur(noisy_img, (5, 5), 0)

# 3. Median Filter
median_filtered = cv2.medianBlur(noisy_img, 5)

# 4. Bilateral Filter
bilateral_filtered = cv2.bilateralFilter(noisy_img, 9, 75, 75)
```

```python
#  Display all filtered images for comparison
show_images(
    [avg_filtered, gaussian_filtered, median_filtered, bilateral_filtered],
    ["Averaging", "Gaussian Blur", "Median Filter", "Bilateral Filter"]
)

#  Compare original with each filter
show_images(
    [noisy_img, avg_filtered],
    ["Original", "Averaging"]
)

show_images(
    [noisy_img, gaussian_filtered],
    ["Original", "Gaussian Blur"]
)

show_images(
    [noisy_img, median_filtered],
    ["Original", "Median Filter"]
)

show_images(
    [noisy_img, bilateral_filtered],
    ["Original", "Bilateral Filter"]
)

# -------------------------------------------------------------

print("Filtering complete. Based on visual comparison:")
print("- Median Filter is usually best for Salt & Pepper noise.")
print("- Gaussian is good for Gaussian noise.")
print("- Bilateral Filter preserves edges very well.")
```
